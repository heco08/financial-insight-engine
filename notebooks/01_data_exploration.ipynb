```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Visual configuration
sns.set(style='darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)

# CSV file path
file_path = os.path.join('..', 'data', 'stock.csv')  # Adjust 'stock.csv' if the filename is different

# Load data
df = pd.read_csv(file_path, parse_dates=['Date']) 

# Show first rows
print("First 5 rows:")
display(df.head())

# Basic statistics
print("\nDescriptive statistics:")
display(df.describe())

# Missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Plot stock prices (adjust column names if necessary)
companies = df.columns.drop('Date')
for company in companies:
    plt.plot(df['Date'], df[company], label=company)

plt.title('Historical Stock Prices')
plt.xlabel('Date')
plt.ylabel('Adjusted Price ($)')
plt.legend()
plt.show()
```
